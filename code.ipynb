```python
# importing libraries
import os
import pandas as pd
import googlemaps
import os
import csv, json
```

```python
# accessing directories
path = "C:/Users/Rahul/Desktop/Task1/complete-towns-directory-indiastatedistrictsub-district-level-census-2011"
files = os.listdir(path)
```

```python
files_ = files[1:35]
```

```python
# reading excel file
s1 = {}
for i in range(len(files_)):
    path1 = files_[i]
    data1 = pd.ExcelFile(path1)
    d = data1.sheet_names[0]
    s1[i] = data1.parse(d)
  
```

```python
# requesting api for access
g_maps = googlemaps.Client(key = 'AIzaSyDFPK2nvsxEIqT8xOZVQ1roLYjN8bXfIIQ')
```

```python
#extracting geocode
for j in range(33,34):
    list_lat =  []
    list_long = []
    for i in range(0,len(s1[j]),1):
        results = g_maps.geocode(s1[j].iat[i,-1])
       
        try:
            lat = results[0]['geometry']['location']['lat']
            lon = results[0]['geometry']['location']['lng']
            print(lat)
        except:
            lat = None
            lon = None
        list_lat.append(lat)
        list_long.append(lon)
    print(list_lat)
    s1[j]['LAT'] = list_lat
    s1[j]['LON'] = list_long

        
```

```python
# storing the data in csv format
mod = {}
for i in range(33,34):
    mod[i] = s1[i].to_csv('submiss' + str(i) + '.csv')
```

```python
# storing the data in excel format
for i in range(33,34):
    writer = pd.ExcelWriter('town' + str(i) + '.xlsx', engine='xlsxwriter')

    # Convert the dataframe to an XlsxWriter Excel object.
    s1[i].to_excel(writer, sheet_name='Sheet1')

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```
